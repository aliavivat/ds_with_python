## 3.1 Pie chart

```python
# Using pandas plot method: 
abalone_df['Height'].plot(kind='hist',bins=30) 
plt.title('Abalone - Height info') 
plt.xlabel('Height') 
plt.grid() 
plt.show() # displays the current figure and closes it 
```

```python
# Using seaborn: 
sns.histplot(data=abalone_df, x='Height',bins=30) 
plt.grid() 
```

## 3.2 Histogram

```python
# Using pandas plot method: 
abalone_df['Height'].plot(kind='hist',bins=30) 
plt.title('Abalone - Height info') 
plt.xlabel('Height') 
plt.grid() 
plt.show() # displays the current figure and closes it 
```

```python
# Using seaborn: 
sns.histplot(data=abalone_df, x='Height',bins=30) 
plt.grid() 
```

## 3.3 Density

```python
# Using pandas plot method: 
abalone_df['Shell weight'].plot(kind='density') 
plt.title('Abalone - Shell Weight') 
plt.xlabel('Shell weight') 
plt.grid() 
plt.show() 
```

```python
# Using seaborn: 
sns.kdeplot(data=abalone_df, x='Shell weight') 
plt.grid() 
```

## 3.4 Boxplot

```python
# Using pandas plot method: 
abalone_df['Whole weight'].plot(kind='box') 
plt.grid() 
plt.show() # this will display  
```

```python
# Using seaborn: 
sns.boxplot(data=abalone_df, y='Whole weight') 
plt.grid() 
```

Draw the boxplot for `Whole weight` per `Sex` group:

```python
# Using pandas boxplot method: 
abalone_df.boxplot(column='Whole weight',by='Sex') 
plt.show() 
```

```python
# Using seaborn: 
sns.boxplot(data=abalone_df, y='Whole weight', x='Sex') 
```

## 3.5 Scatterplot

```python
# Using pandas plot method: 
abalone_df.plot(kind='scatter',x=8,y=2,figsize=(8,6)) 
plt.grid() 
plt.show() 
```

```python
# Using seaborn: 
plt.figure(figsize=(8,6)) 
sns.scatterplot(data=abalone_df, x='Rings', y='Diameter') 
plt.grid() 
```

## 3.6 Scatter matrix

```python
from pandas.plotting import scatter_matrix 
# Using pandas scatter matrix function: 
scatter_matrix(abalone_df, alpha=0.2, figsize=(16,16), diagonal='hist') 
```

```python
# Using seaborn: 
sns.pairplot(abalone_df, diag_kind='hist') 
```
